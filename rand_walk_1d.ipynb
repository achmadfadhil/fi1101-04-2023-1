```python
# ===========================================
# Random Walk 1D
# ===========================================
# Name: Ahmad Fadillah
# SID : 10223006
# File: rand_walk_1d.ipynb
# ===========================================

import numpy as np
import matplotlib.pyplot as plt

# -------------------------------------------------
# Fungsi Simulasi Random Walk 1D
# -------------------------------------------------
def random_walk_1d(Nrep=1000, Nstep=200):
    """
    Simulasi random walk satu dimensi (1D).
    Nrep  : jumlah pengulangan (repetisi)
    Nstep : jumlah langkah tiap partikel
    Return: mean posisi  dan mean kuadrat posisi
    """
    # Setiap langkah bernilai +1 (kanan) atau -1 (kiri)
    steps = np.random.choice([-1, 1], size=(Nrep, Nstep))
    # Akumulasi posisi tiap langkah
    positions = np.cumsum(steps, axis=1)
    # Rata-rata posisi dan posisi kuadrat
    mean_x = np.mean(positions, axis=0)
    mean_x2 = np.mean(positions**2, axis=0)
    return mean_x, mean_x2


# -------------------------------------------------
# Parameter simulasi
# -------------------------------------------------
Nstep = 200
Nreps = [100, 1000, 10000]  # variasi jumlah ulangan

# -------------------------------------------------
# Visualisasi Mean Distance
# -------------------------------------------------
plt.figure(figsize=(7,4))
for Nrep in Nreps:
    mean_x, _ = random_walk_1d(Nrep, Nstep)
    plt.plot(mean_x, label=f'Nrep={Nrep}')
plt.xlabel('Steps')
plt.ylabel('Mean Distance ⟨x⟩')
plt.title('1D Random Walk — Mean Distance')
plt.grid(True)
plt.legend()
plt.show()

# -------------------------------------------------
# Visualisasi Mean Squared Distance
# -------------------------------------------------
plt.figure(figsize=(7,4))
for Nrep in Nreps:
    _, mean_x2 = random_walk_1d(Nrep, Nstep)
    plt.plot(mean_x2, label=f'Nrep={Nrep}')
plt.xlabel('Steps')
plt.ylabel('Mean Squared Distance ⟨x²⟩')
plt.title('1D Random Walk — Mean Squared Distance')
plt.grid(True)
plt.legend()
plt.show()

# -------------------------------------------------
# Analisis hasil simulasi (ringkasan numerik)
# -------------------------------------------------
mean_x, mean_x2 = random_walk_1d(Nrep=10000, Nstep=200)
print("Langkah terakhir (Nstep = 200):")
print(f"⟨x⟩  = {mean_x[-1]:.3f}")
print(f"⟨x²⟩ = {mean_x2[-1]:.3f}")

```

```python
# Random Walk 1D
Simulasi ini menunjukkan perilaku **langkah acak satu dimensi (1D random walk)**.
Tujuan:
- Menghitung dan menampilkan *mean distance* (⟨x⟩)
- Menghitung *mean squared distance* (⟨x²⟩)
- Menganalisis pengaruh banyaknya ulangan simulasi (Nrep)
```

```python
## Analisis Tugas 1 dan 2

- Hasil simulasi menunjukkan bahwa **mean distance ⟨x⟩** selalu mendekati nol, konsisten dengan teori bahwa peluang langkah kiri dan kanan sama besar.
- **Mean squared distance ⟨x²⟩** meningkat hampir linear terhadap jumlah langkah, sesuai hubungan ⟨x²⟩ = n.
- Peningkatan jumlah ulangan (Nrep) menyebabkan grafik menjadi lebih halus karena fluktuasi statistik menurun (~1/√Nrep).
- Secara umum hasil ini konsisten dengan teori *Brownian motion*.

**Referensi:**
- Feller, W. *An Introduction to Probability Theory and Its Applications*, Vol. 1, Wiley.
- Einstein, A. (1905). *On the movement of small particles suspended in stationary liquids required by the molecular-kinetic theory of heat.*
```
